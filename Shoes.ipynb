## Data Cleansing

```python
df['Size'] = df['Size'].str.replace('US ', '').astype(float)
```

```python
df['Price (USD)'] = df['Price (USD)'].str.replace('$', '').astype(float)
```

## Data Visualizations



### Visualizations involving **'Size'**

```python
from matplotlib.ticker import MultipleLocator


sns.histplot(df['Size'], color = 'purple', bins = range(5, 13), binwidth= 0.5, ec = 'black');
plt.gca().xaxis.set_major_locator(MultipleLocator(0.5))
plt.title('Frequency of Shoe Sizes')
plt.xlabel('Size')
plt.ylabel('Frequency');

```

```python

mean_prices = df.groupby('Size')['Price (USD)'].mean()

plt.scatter(mean_prices.index, mean_prices.values);
plt.title('Average Shoe Price by Size')
plt.xlabel('Shoe Size')
plt.ylabel('Price (USD)');
plt.gca().xaxis.set_major_locator(MultipleLocator(0.5))
```

```python
sns.boxplot(x=df['Size'], y=df['Price (USD)']);
plt.xlabel('Size')
plt.ylabel('Price (USD)')
plt.title('Price Distribution by Size');
```

```python
# Group data by shoe size and gender and calculate mean price
grouped_data = df.groupby(['Size', 'Gender'])['Price (USD)'].mean().reset_index()

# Create bar plot
sns.barplot(x='Size', y='Price (USD)', hue='Gender', data=grouped_data)
plt.title('Average Shoe Prices by Size and Gender')
plt.xlabel('Shoe Size')
plt.ylabel('Average Price (USD)');
```

```python
table = pd.pivot_table(df, values='Price (USD)', index='Brand', columns='Size', aggfunc='count')

sns.heatmap(table, cmap='Blues', annot=True, fmt='g')
plt.title('Brands by Shoe Size');
```

### Visualizations involving **'Brand'**

```python
brand_counts = df['Brand'].value_counts()
sns.barplot(x=brand_counts.index, y=brand_counts.values, order=brand_counts.index, color='b')
plt.title('Count of Shoes by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45);
```

```python
# Group data by brand and gender and count number of shoes
grouped_data = df.groupby(['Brand', 'Gender'])['Size'].count().reset_index()

# Pivot the data to create a stacked bar chart
pivot_data = grouped_data.pivot(index='Brand', columns='Gender', values='Size')

# Create stacked bar chart
pivot_data.plot(kind='bar', stacked=True)
plt.title('Number of Shoes by Brand and Gender')
plt.xlabel('Brand')
plt.ylabel('Number of Shoes')
plt.xticks(rotation=45);
```

```python
sns.boxplot(x='Brand', y='Price (USD)', data=df)
plt.xticks(rotation=45);
plt.title('Price Distribution by Brand');
```

```python
# Pivot the data to create a heatmap
heatmap_data = df.pivot_table(index='Brand', columns='Type', values='Size', aggfunc='count')

# Create the heatmap
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='g')
plt.title('Shoe Types by Brand');
```

```python
# Group data by brand and count number of shoe types
grouped_data = df.groupby('Brand')['Type'].nunique().reset_index()

# Sort data by number of shoe types
grouped_data = grouped_data.sort_values('Type', ascending=False)

# Create bar chart
plt.bar(grouped_data['Brand'], grouped_data['Type'])
plt.xticks(rotation = 45)
plt.xlabel('Brand')
plt.ylabel('Number of Shoe Types')
plt.title('Number of Shoe Types by Brand')
plt.show()
```

```python
# Create a subset of data for the bubble chart
subset = df[['Brand', 'Type', 'Price (USD)']].dropna()

# Group data by brand and type and calculate the average price
grouped_data = subset.groupby(['Brand', 'Type'])['Price (USD)'].mean().reset_index()

# Create the bubble chart
plt.figure(figsize=(12,6))
sns.scatterplot(x='Type', y='Brand', size='Price (USD)', data=grouped_data, sizes=(50, 1000), alpha=0.8)
plt.xlabel('Type of Shoe')
plt.ylabel('Brand')
plt.title('Average Price by Brand and Type')
plt.xticks(rotation=45);
```

### Visualizations Involving **'Model'**

```python
from wordcloud import WordCloud

model_text = ' '.join(list(df['Model']))

wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(model_text)

plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
```

```python
model_counts = df['Model'].value_counts().nlargest(10)

plt.bar(model_counts.index, model_counts.values)
# Create bar chart
plt.xticks(rotation = 60)
plt.xlabel('Model')
plt.ylabel('Number of Shoe Models')
plt.title('10 Most Common Shoe Models')
plt.show()
```

```python
# Group data by model and get the mean price
model_mean_price = df.groupby('Model')['Price (USD)'].mean()

# Sort by price
sorted_model_price = model_mean_price.sort_values(ascending=False)

# Get the top 5 and bottom 5 models by price
top_5 = sorted_model_price.head(5)
bottom_5 = sorted_model_price.tail(5)

# Create a bar chart of the top 5 and bottom 5 models
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(top_5.index, top_5.values)
plt.xticks(rotation=45)
plt.title('Top 5 Most Expensive Models')
plt.xlabel('Model')
plt.ylabel('Price')

plt.subplot(1,2,2)
plt.bar(bottom_5.index, bottom_5.values)
plt.xticks(rotation=45)
plt.title('Top 5 Least Expensive Models')
plt.xlabel('Model')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

## Visualizations Involving **'Type'**

```python
type_counts = df['Type'].value_counts()

plt.bar(type_counts.index, type_counts.values);
plt.xticks(rotation=60);
plt.xlabel('Type of Shoe')
plt.ylabel('Count')
plt.title('Shoe Counts by Type')

# adds data labels for each bar
for i, v in enumerate(type_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
```

```python
avg_price_by_type = df.groupby('Type')['Price (USD)'].mean().reset_index()


# Sort the bars in descending order
avg_price_by_type.sort_values(by='Price (USD)', ascending=False, inplace=True)

plt.bar(avg_price_by_type['Type'], avg_price_by_type['Price (USD)'])
plt.xticks(rotation=90)
plt.title('Average Price by Type');
plt.xlabel('Type of Shoe')
plt.ylabel('Average Price');
```

```python
# Subset the data by gender
df_men = df[df['Gender'] == 'Men'];
df_women = df[df['Gender'] == 'Women'];

# Calculate the average price by type for men and women separately
avg_price_men = df_men.groupby('Type')['Price (USD)'].mean();
avg_price_women = df_women.groupby('Type')['Price (USD)'].mean();

# Combine the two dataframes
avg_price_by_type = pd.concat([avg_price_men, avg_price_women], axis=1);
avg_price_by_type.columns = ['Men', 'Women'];

# Plot the grouped bar chart
ax = avg_price_by_type.plot(kind='bar', width=0.8, figsize=(12,6), color=['#2c7fb8', '#f03b20'], edgecolor='black');

# Set the title and axis labels
ax.set_title('Average Price by Type and Gender');
ax.set_xlabel('Type');
ax.set_ylabel('Average Price');

# Add data labels
for i, v in enumerate(avg_price_by_type['Men']):
    ax.text(i - 0.2, v + 10, f"${float(v):.0f}", ha='center', fontsize=10);
for i, v in enumerate(avg_price_by_type['Women']):
    ax.text(i + 0.4, v + 15, f"${float(v):.0f}", ha='center', fontsize=10); 
# Add a legend
ax.legend(loc='upper right', frameon=True);

```

## Visualizations Involving **'Color'**


```python
# Set the threshold value for minimum number of shoes in each color
threshold = 10

# Count the number of shoes for each color
color_counts = df['Color'].value_counts()

# Get the colors that meet the threshold value
main_colors = color_counts[color_counts >= threshold].index.tolist()

# Calculate the count of shoes in the 'Other' category
other_count = color_counts[color_counts < threshold].sum()

# Group the colors that don't meet the threshold value into an "Other" category
color_counts = color_counts[color_counts >= threshold]
color_counts.loc['Other'] = other_count

# Create the bar chart
plt.figure(figsize=(12,8))
sns.barplot(x = color_counts.index, y = color_counts.values)
plt.title('Count of Shoes by Color Where Count > 10')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=90);
```

```python
# Set the threshold value for minimum number of shoes in each color
threshold = 10

# Count the number of shoes and average price for each color
color_stats = df.groupby('Color').agg({'Price (USD)': 'mean', 'Brand':'count'})
color_stats = color_stats[color_stats['Brand'] >= threshold].sort_values('Price (USD)', ascending=False).head(15)

# Create the bar chart
plt.figure(figsize=(12,8))
sns.barplot(x=color_stats.index, y = color_stats['Price (USD)'], order = color_stats.index)
plt.title('Average Price of Shoes by Color (Top 15)')
plt.xlabel('Color')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=90);
```

```python

# Set the threshold value for minimum number of shoes in each color
threshold = 10

# Filter the data to colors that meet the threshold value
color_counts = df['Color'].value_counts()
main_colors = color_counts[color_counts >= threshold].index.tolist()
df = df[df['Color'].isin(main_colors)]


# Calculate the percentage of each color for each type
color_type_counts = df.groupby(['Type', 'Color']).size().reset_index(name='Count')
color_totals = color_type_counts.groupby('Type')['Count'].sum().reset_index(name='Total')
color_type_counts = pd.merge(color_type_counts, color_totals, on='Type')
color_type_counts['Percentage'] = color_type_counts['Count'] / color_type_counts['Total']

# Create the heatmap with a blue color scheme
plt.figure(figsize=(18, 14))
ax = sns.heatmap(color_type_counts.pivot(index='Color', columns='Type', values='Percentage'), cmap='Blues')
plt.title('Likelihood of Most Common Colors by Shoe Type', fontsize = 15)

# Add data labels
for i, (rowname, row) in enumerate(color_type_counts.pivot(index='Color', columns='Type', values='Percentage').iterrows()):
    for j, value in enumerate(row):
        if not np.isnan(value):
            ax.text(j + 0.5, i + 0.5, f'{value:.2%}', ha='center', va='center')
```

## Visualizations Involving **'Material'**

```python
# Set the threshold value for minimum number of shoes made of each material
threshold = 5

# Count the number of shoes for each material
material_counts = df['Material'].value_counts()

# Get the materials that meet the threshold value
main_materials = material_counts[material_counts >= threshold].index.tolist()

# Calculate the count of shoes made of materials that don't meet the threshold value
other_count = material_counts[material_counts < threshold].sum()

# Group the materials that don't meet the threshold value into an "Other" category
material_counts = material_counts[material_counts >= threshold]
material_counts.loc['Other'] = other_count

# Create the bar chart
plt.figure(figsize=(12,8))
sns.barplot(x = material_counts.index, y = material_counts.values)
plt.title('Count of Shoes by Material')
plt.xlabel('Material')
plt.ylabel('Count')
plt.xticks(rotation=45);


# Add data labels to the bars
for i, v in enumerate(material_counts.values):
    plt.text(i, v+10, str(v), ha='center', fontsize=10)
```

```python
# Create a pivot table of material vs. size
pivot_table = pd.pivot_table(df, values='Size', index='Material', columns='Brand', aggfunc='count')

# Create the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt='g')
plt.title('Frequency of Materials Used in Different Sizes')
plt.xlabel('Size')
plt.ylabel('Material')
plt.show()
```

```python
# Group the data by material and calculate the mean price for each group
material_prices = df.groupby('Material')['Price (USD)'].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(20,12))
sns.barplot(x=material_prices.index, y=material_prices.values)
plt.title('Average Price by Material', fontsize = 28)
plt.xlabel('Material', fontsize = 25)
plt.ylabel('Average Price', fontsize = 20)
plt.xticks(rotation=90, fontsize = 20);
plt.yticks(fontsize = 15);


```
